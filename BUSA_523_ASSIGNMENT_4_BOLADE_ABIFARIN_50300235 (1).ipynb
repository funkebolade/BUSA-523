```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("data.csv",header = 0, index_col = 0)
```

```python
data.head()
```

```python
data.dropna()
```

### (Q1a) HISTOGRAM

```python
##for Temperture
```

```python
plt.figure(figsize=(14,6))
bins = [40,50,60,70,80,90,100]
plt.hist(data.dropna()["Temperature"], bins = bins)
plt.xticks(bins)
plt.xlabel("Temperature")
plt.ylabel("frequency")
plt.title("WEATHER")

```

```python
###for CPI
```

```python
plt.hist(data.dropna()["CPI"]) 
plt.xlabel("CPI")
plt.ylabel("frequency")
plt.title("CONSUMPTION")
```

```python
###For Unemployment
```

```python
plt.hist(data.dropna()["Unemployment"]) 
plt.xlabel("Unemployment")
plt.ylabel("frequency")
plt.title("UNEMPLOYMENT RATE")
```

### (Q1b) BAR CHART

```python
## For the whole data
```

```python

data.dropna().plot.bar()
```

```python
data.dropna().plot.barh()
```

```python
## For Fuel price and sales
```

```python
plt.figure(figsize=(14,5))
w = 0.4
x_1 = np.arange(len(data.dropna()))
y_2 = [i + w for i in x_1]
fuel_price = data.dropna()["Fuel_Price"]
dept_1 = data.dropna()["dept_1 sales"]

plt.bar(x_1, fuel_price,label = "Fuel")
plt.bar(y_2, dept_1,w ,label = "dept_")

plt.xlabel("dept_ 1")
plt.ylabel("fuel price")
plt.title("SALES")
plt.legend()
```

```python
plt.figure(figsize=(14,5))
w = 0.4
x_1 = np.arange(len(data.dropna()))
y_2 = [i + w for i in x_1]
fuel_price = data.dropna()["Fuel_Price"]
dept_1 = data.dropna()["dept_1 sales"]

plt.barh(x_1, fuel_price,label = "Fuel")
plt.barh(y_2, dept_1,w ,label = "dept_")

plt.xlabel("dept_ 1")
plt.ylabel("fuel price")
plt.title("SALES")
plt.legend()
```
